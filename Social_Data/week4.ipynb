```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import folium
from scipy import stats
import matplotlib.pyplot as plt
from datetime import datetime

# Load dataset 
data = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")

data.head()
```

```python
sub_data = data[data.Category == 'DRUG/NARCOTIC']
sub_set = sub_data[((pd.DatetimeIndex(sub_data['Date']).month == 6) | (pd.DatetimeIndex(sub_data['Date']).month == 7))
                   & (pd.DatetimeIndex(sub_data['Date']).year == 2016)]
```

```python
map_SF = folium.Map(location=[37.733795, -122.446747],tiles = "Stamen Terrain", vzoom_start = 13)
folium.Marker([37.77919, -122.41914], popup='City Hall',icon=folium.Icon(color='blue')).add_to(map_SF)
for idx, row  in sub_set.iterrows():
    folium.Marker([row['Y'],row['X']], popup = 'Drug/Narcotics').add_to(map_SF)
map_SF
```

```python
data_sex = data[data.Category == 'SEX OFFENSES, NON FORCIBLE']
data_sex = data_sex[((pd.DatetimeIndex(data_sex['Date']).month == 6) | (pd.DatetimeIndex(data_sex['Date']).month == 7))
                   & (pd.DatetimeIndex(data_sex['Date']).year == 2016)]
```

```python
from folium.plugins import HeatMap
map_SF = folium.Map(location=[37.733795, -122.446747], vzoom_start = 13)

# Ensure you're handing it floats
data_sex['Y'] = data_sex['Y'].astype(float)
data_sex['X'] = data_sex['X'].astype(float)

# Create weight column, using date
heat_df = data_sex[['Y','X']]
heat_df = heat_df.dropna(axis=0, subset=['Y','X'])

# List comprehension to make out list of lists
heat_data = [[row['Y'],row['X']] for index, row in heat_df.iterrows()]

# Plot it on the map
HeatMap(heat_data).add_to(map_SF)
map_SF
```

```python
data_sex = data[data.Category == 'DRUG/NARCOTIC']
```

```python
from folium import plugins
map_SF = folium.Map(location=[37.733795, -122.446747], vzoom_start = 13)

heat_df = data_sex[(pd.DatetimeIndex(data_sex['Date']).month == 6)]

# Ensure you're handing it floats
data_sex['Y'] = data_sex['Y'].astype(float)
data_sex['X'] = data_sex['X'].astype(float)

# Create weight column, using date
heat_df['Weight'] = heat_df['Date'].str[3:5]
heat_df['Weight'] = heat_df['Weight'].astype(float)
heat_df = heat_df.sort_values(by='Weight', ascending = True)

heat_df = heat_df.dropna(axis=0, subset=['Y','X','Weight'])

# List comprehension to make out list of lists
heat_df = [[[row['Y'],row['X']] for index, row in heat_df[heat_df['Weight'] == i].iterrows()] for i in range(0,13)]

# Plot it on the map
hm = plugins.HeatMapWithTime(heat_df,auto_play=True,max_opacity=0.8)
hm.add_to(map_SF)
map_SF
```
