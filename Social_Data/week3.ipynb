So this time, the distribution we want to plot is the average time-of-day that a crime takes place. There are many ways to do this, but let me describe one way to do it. 
    * For datapoint, the only thing you care about is the time-of-day, so discard everything else.
    * You also have to deal with the fact that time is annoyingly not divided into nice units that go to 100 like many other numbers. I can think of two ways to deal with this.
      * For each time-of-day, simply encode it as seconds since midnight.
      * Or keep each whole hour, and convert the minute/second count to a percentage of an hour. So 10:15 $\rightarrow$ 10.25, 8:40 $\rightarrow$ 8.67, etc.

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from datetime import datetime

# Load dataset 
data = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")

data.head()
```

```python
sub_data = data[data.Category == 'PROSTITUTION']
```

```python
sub_set = sub_data[((pd.DatetimeIndex(sub_data['Date']).month == 1) | (pd.DatetimeIndex(sub_data['Date']).month == 2) |(pd.DatetimeIndex(sub_data['Date']).month == 3))
                   & (pd.DatetimeIndex(sub_data['Date']).year == 2007)]
sub_set = sub_set[(pd.DatetimeIndex(sub_set['Time']).hour ==20)]
sub_set = sub_set.reset_index()
# Jitters plot 
plt.plot(sub_set['Time'].sort_values(), 'o',color = 'black', mfc = 'none');
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
ax = sns.stripplot(x=sub_set['Time'],jitter=0.1,edgecolor='none', alpha=.90)
```
