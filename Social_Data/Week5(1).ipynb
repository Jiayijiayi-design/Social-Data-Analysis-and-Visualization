```python
import matplotlib.pyplot as plt
from matplotlib.image import NonUniformImage
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv")

# remove outliers: 
data = data[(data.Y < 50) & (data.X<-122)]
```

```python
# log plot
fig = plt.figure(figsize = (16, 8))
plt.yscale("log")
data['Category'].value_counts().plot(kind='bar',title='Number of crimes in each category')
```

```python
# loglog plot.
# Step 1: Divide San Francisco into a grid roughly 100m x 100m
from geopy import distance
import math 

lat_max = np.max(data['X'])
lat_min = np.min(data['X'])

lo_max = np.max(data['Y'])
lo_min = np.min(data['Y'])

d1 = (lo_min, lat_min)
d2 = (lo_min, lat_max)
lat_dis = distance.distance(d1 , d2 ).m

d3 = (lo_max, lat_max)
lo_dis = distance.distance(d2 , d3 ).m

print(lo_dis/100)
print(lat_dis/100)
```

```python
H, xedges, yedges  = np.histogram2d(data['X'], data['Y'], bins=[math.floor(lo_dis/100), math.floor(lat_dis/100)]) 
H = H.T
```

```python
plt.figure(figsize=(10,10))
plt.imshow(H, cmap='hot',vmax=3000, origin='lower')
plt.title('Total number of crimes in San Francisco')
plt.colorbar()
```

```python
data.Category.unique()
```

```python
# Step 2: Count the number of thefts occurring within each grid-square (use all data for all time).
thefts = ['LARCENY/THEFT', 'VEHICLE THEFT']
Subset_theft = data[data["Category"].isin(thefts)]
Subset_theft.head()
```

```python
H_theft, _, _  = np.histogram2d(Subset_theft['X'], Subset_theft['Y'], bins=[math.floor(lo_dis/100), math.floor(lat_dis/100)]) 
H_theft = H_theft.T
plt.figure(figsize=(10,10))
plt.imshow(H_theft, cmap='hot',vmax=3000, origin='lower')
plt.title('Total number of crimes in San Francisco')
plt.colorbar()
```

```python
Subset_theft.shape
```

Step 3: Tally the counts. Count the number of squares with $k=0$ thefts. We call this $N(0)$. Next, count the number of grids with one crime to get $N(k=1)$. Keep going like this all the way up to $k=C_{max}$, where $C_{max}$ is the highest count of crimes you find in any grid space. 
>     * *Extra tip*: If you want all the details on binning for loglog axes, you can check out [Lecture 2, exercise 3]

```python
# create 1.D array of the Histogram bin count
H_1D=H_theft.ravel()
# make the N that counts for each value k up to max.
N = []
t = range(np.max((H_1D+1).astype(int)))
N = [np.count_nonzero(H_1D==ii) for ii in t]
```

>   - Step 4: Plot the distribution of $k+1$ vs $N(k)$ on linear axes.

```python
fig, ax = plt.subplots()
ax.plot(t, N)

ax.set(xlabel='k', ylabel='N(k)',
       title='loglog plot of N(k)')
ax.grid()
plt.show()
```

```python
## simple plot graph of the k against N
fig, ax = plt.subplots()
ax.plot(t, N)

ax.set(xlabel='k', ylabel='N(k)',
       title='loglog plot of N(k)')
ax.grid()
plt.yscale("log")
plt.yscale("log")
plt.show()
```

Next it's time for the linear regression. Janert writes about this on page 63-66.

There is a closed-form solution for linear regression. If you want to find the best straight line $y = ax + b$ fit to a set of $N$ points $\{(x_1,y_1), (x_2,y_2), \ldots (x_N,y_N)\}$, the value of $b$ is

$$
\tag{1}
b = \langle y \rangle - a \langle x \rangle,
$$

where $\langle x \rangle = (1/N)\sum_i x_i$ is the mean value of the $x_i$ and $\langle y \rangle = (1/N)\sum_i x_i$ is the mean value of the $y_i$. 

And the value for the slope $a$ is 

$$
\tag{2}
a = \frac{\sum_{i=1}^N \left( x_iy_i \right) - N \langle x\rangle\langle y\rangle }{\sum_{i=1}^N\left( x_i^2 \right) - N\langle x\rangle^2}.
$$

Last year, I actually derived this whole thing. I've taken it out of the notebook. But if you'd like to take a look (it's a fun and instructive little exercise), you can find it **[here](https://github.com/suneman/socialdata2021/blob/main/lectures/LinearRegressionDerived.ipynb)**.

### Goodness of fit as a measure of correlation

But the question we started with was "Which pair of focus crimes have the the most similar temporal pattern across the week?". We haven't really answered that yet. So let's calculate one last thing: $R^2$. You probably also remember this one. 

Basically $R^2$ is a measurea of how good a linear fit is. You can [read about $R^2$ on wikipedia](https://en.wikipedia.org/wiki/Coefficient_of_determination).

In statistics, $R^2$ is a value that measures how much variance of the dependent variable (y) is explained by the independent variables (x). $R^2$ ranges from 0 to 1. 
The expression is 

$R ^2  = 1 - SS_{res}/ SS_{tot}$

Where the $SS_{tot}$ is the variance in the observed data, and the $SS_{res}$ is the residuals, which is a measure of how good the predicted model is. 
This value have not so much to do with pearson coefficient, it is a measure of how good the model is. 
This $R^2$ is correlation between prediction and true output. 

```python
R2=np.zeros(91)

for index, row in LinearReg.iterrows():
    
    x = data[data.Category==row['x']].hours.value_counts().sort_index()
    y = data[data.Category==row['y']].hours.value_counts().sort_index()
    
    f = row['a']*x + row['b']
    
    SSres=np.sum((y-f)**2)
    SStot= np.sum((y-np.mean(y))**2)
    
    R2[index]=1-(SSres/SStot)
    
LinearReg['R2']=R2
```
